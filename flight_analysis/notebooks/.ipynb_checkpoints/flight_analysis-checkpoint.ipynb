## Data Cleaning - BTS

```python
#finds every .csv file in the bts folder and returns a list of their paths
files = glob.glob("../../data/bts/*.csv")

df = pd.concat([pd.read_csv(f) for f in files], ignore_index = True)

print(df.columns.tolist())

#convert date from object to datetime
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], format='mixed')

#create new columns for day and month
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.day_name

df['MONTH'] = df['FL_DATE'].dt.month

print(df.dtypes)

#check for any missing values - maybe in delay columns that could be empty

#fix any data types that are incorrect



```
