### Questions about the data  

Which airlines have the best/worst on-time performance? 

- This helps travelers make more informed booking decisions and enables airlines to see where they stand against competitors.

Which airlines have the best cost-to-performance ratio?

- This benefits budget-conscious travelers by identifying airlines that are reliable and aren't overly pricey. 

Which airports cause the biggest delays across networks? (which are the biggest bottlenecks)

- This can help airlines optimize routing to limit the spread of delays throughout the flight network.

What are common causes of delays?

- Airlines and airports can understand root causes for delays and target where improvements will be the most beneficial/

Are certain routes worse than others?

- This could give airlines insights into scheduling readjustments and inform travelers of problematic routes to be aware of before booking.

Do delays vary by season?

- Recognizing seasonal patterns can help travelers plan around high-risk travel periods, and help airlines to optimize staffing and scheduling
