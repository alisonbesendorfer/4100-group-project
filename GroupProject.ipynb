Data Source #1: Bureau of Transportation Statistics - Performance Data

- This is our primary data source. It is available to us through direct download from the  BTS Reporting Carrier On-Time Performance database at *https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGJ&QO_fu146_anzr=b0-gvzr*. The data is in CSV format and must be downloaded on a month-to-month timeline, and we can select specific fields for each download. Each row represents a single flight, and we can narrow it down to only flights with delays. Key attributes include: reporting airline, flight number, origin and destination airport codes, scheduled and actual departure/arrival times, delay duration in minutes, cause of delay, cancellation, and distance. This is publicly available U.S. government data with no account required. The only downside is that we have to repeadtly queue downloads.

Data Source #2: National Oceanic and Atmospheric Administration - Historical Weather Observations

- Our secondary dataset comes from the National Oceanic and Atmospheric Administration's Climate Data Online service. The API documentation is available at *https://www.ncdc.noaa.gov/cdo-web/webservices/v2*. Access requires a free API token, which can be requested at *https://www.ncdc.noaa.gov/cdo-web/token*. We will gather daily weather summaries for weather stations located at or near the major airports in our flight data. Attributes include daily temperature highs and lows, precipitation, snowfall, wind speed, and visibility. The data is available in JSON format. The challenge here is matching NOAA weather station IDs to FAA airport codes and aligning daily weather observations with individual flight records by date and location.

Data Source #3: FAA Airport Facilities Data

- In order to enrich our analysis with airport metadata, we will use the FAA's airport facilities data available at *https://adip.faa.gov/agis/public/#/airportSearch/advanced*. This provides attributes such as airport latitude/longitude, elevation, number of runways, airport ownership type, and hub classification (large, medium, small, non-hub). This data is publicly available through direct download and does not require an account. It is in a structured tabular format. This source allows us to ask questions about whether airport characteristics like size, hub status, or geographic location correlate with delay frequency.

Stakeholders who may be affected by this project and its outcomes:

Airlines:

- Our analysis could rank airlines by on-time performance and cost-to-performance ratios. Publishing these findings could influence public perception of specific airlines. If our methodology is flawed, we could unfairly damage an airline's reputation.

Travelers/Consumers:

- Our findings could influence booking decisions. If travelers rely on our analysis to avoid certain airlines or routes, and our conclusions are based on incomplete data or a limited time window, we could be steering people toward worse decisions rather than better ones. Budget-conscious travelers in particular could be affected, if we were to label their preferred airline as unreliable

Airports:

- Identifying specific airports as poor performers in terms of delays could affect an airport's reputation, which in turn affects local economies that depend on air traffic and tourism. When looking at smaller airports, our reporting could disproportionately affect them, as their resources could be more finite, and they face structural disadvantages compared to larger hubs.

How could this project be used for harm?

- Our analysis could be taken out of context or oversimplified. For instance, labeling a carrier as the "worst" based on delay averages alone ignores a multitude of other factors. For example, an airline that operates heavily out of weather-prone hubs like Chicago O'Hare or Minneapolis will naturally have higher delay rates than one based in a mild climate such as Phoenix, and that does not reflect operational issues. If our findings were cited without this context, it could mislead consumers or harm airlines. There is also a risk of reinforcing existing disparities. Budget carriers tend to serve more frugal travelers and if our analysis frames budget airlines negatively without acknowledging the tradeoffs that make affordable air travel possible, it could discourage people from flying with those airlines and opting not to travel alltogether. Finally, weather data tied to geographic locations raises minor privacy concerns. While individual flights are not tied to passenger identities in our data, consistently highlighting that certain regions experience more delays could have unintended economic consequences. For example, it could turn away  business investment or tourism in cities that are marked as places with high delays.

Because we are merging three separate datasets, each with its own formatting and features, a significant amount of our work will be dedicated to preprocessing. For the BTS flight data, we need to put plenty of separate monthly CSV files into a single dataframe. In the BTS flight data, we expect null arrival times for cancelled flights and inconsistent formatting, among other issues. We will match stations to airports using coordinates. We also expect missing observations for certain days or variables and will need to decide how to handle those gaps. For the FAA airport data, we will extract/standardize fields like hub classification, runway count, and coordinates. Beyond cleaning, we plan to create several new features from the raw data, such as average delay by airport per month. Data processing will be implemented Python and the pandas package.

## Exploratory Analysis
Which methods and visualizations are you planning to use to look at your dataset?

- For our exploratory analysis, we plan to use a combination of statistical methods and visualizations to understand the structure of our data. We will start with summary statistics and distributions. Using histograms and box plots, we will examine the distribution of delay times. We expect delay distributions to be heavily right-skewed, with most flights experiencing little to no delay and a long tail of severe delays. Next, we will build a correlation matrix and visualize it as a heat map. This will let us quickly identify which variables have strong linear relationships with delay duration. It will also help us spot multicollinearity between predictor variables, which is important to address before fitting regression models. For geographic patterns, we will use scatter plots or map-based visualizations to display delay frequency and severity across airports in the United States. This can reveal whether delays cluster in certain regions, such as the Northeast or Midwest during the winter. We also plan to use scatter plots with regression lines to explore individual relationships, such as delay duration versus precipitation or delay duration versus number of daily departures at an airport. These plots will help us look at the strength and form of relationships before fitting more complex models.
